```python
# importing pandas package
import pandas as pd
import matplotlib.pyplot as plt 
# making data frame from csv file
data = pd.read_csv("AmesHousingGeo.csv")
data.head()
```

```python
x = data.lotarea
y = data.overallcond

plt.title('Lot Area vs Overall Condition')
plt.xlabel('Lot Area')
plt.ylabel('Overall Condition')
plt.plot(x,y)
plt.show()
```
