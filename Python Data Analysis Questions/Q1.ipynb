```python
#This code converts the .dta file to a readable .csv file
# importing pandas package
import pandas as pd
import matplotlib.pyplot as plt 
# making data frame from csv file
data = pd.read_csv("ekonomiks.csv")
data.head()
```

```python
x = data.pages
#this refers to the productivity
y = data.salary

plt.title('Productivity')
plt.xlabel('Pages')
plt.ylabel('Salary')
plt.plot(x,y)
plt.show()
```

```python
x = data.gender
#this refers to the gender
y = data.salary

plt.title('Productivity per Gender')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.plot(x,y)
plt.show()
```

```python
x = data.phd_rank
#this refers to the phd rank
y = (data.year - data.phd_awarded) 

plt.title('Relevance of PhD Program')
plt.xlabel('PhD Rank')
plt.ylabel('No of Years Before Attaining PhD')
plt.plot(x,y)
plt.show()
```
